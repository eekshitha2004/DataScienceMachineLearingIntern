```python
# 1. Read the csv file named windpower.csv
import pandas as pd
df = pd.read_csv("windpower.csv")
print(df)
```

```python
# 2. Print the top 10 rows
df.head(10)
```

```python
# 3. Print the bottom 5 rows
df.tail()
```

```python
# 4. Display the structure of dataset with datatypes.
df.info()
```

```python
# 5. Find out Mean, Standard deviation, Min, Max for the column wind speed
print("Mean = ",df["Wind Speed (m/s)"].sum())
print("Standard deviation = ",df["Wind Speed (m/s)"].std())
print("Minimum = ",df["Wind Speed (m/s)"].min())
print("Maximum = ",df["Wind Speed (m/s)"].max())
```

```python
# 6.Fill the NULL values with the average value of that column
print(df["Wind Speed (m/s)"].fillna(df["Wind Speed (m/s)"].mean(),inplace = True))
print(df.info())
```

```python
# 7. Create a histogram plot for the column windspeed
import matplotlib.pyplot as plt
df["Wind Speed (m/s)"].plot(kind="hist")
plt.show()
```

```python
# 8. Create scatter plot between windspeed and Theoretical power curve.
df.plot(kind="scatter",x="Wind Speed (m/s)",y="Theoretical_Power_Curve (KWh)")
plt.show()
```

```python
# 9. Display the rows where windspeed <5.
print(df[df["Wind Speed (m/s)"] < 5].to_string())
```

```python
# 10. Extract the rows where wind direction >200
print(df[df["Wind Direction (°)"] > 200].to_string())
```
