```python
import pandas as pd

df = pd.read_csv("insurance.csv")
print(df.to_string())
```

```python
#1. Remove the rows which is having NULL Values.
new_df = df.dropna()
print(new_df.to_string())
```

```python
#replacing with mean
x1 = df.charges.mean()
new_df1 = df.fillna(x1)
print(new_df1.to_string())
```

```python
#replacing with median
x2 = df.charges.median()
new_df2 = df.fillna(x2)
print(new_df2.to_string())
```

```python
#replacing with mode
x1 = df.charges.mode()[0]
new_df1 = df.fillna(x1)
print(new_df1.to_string())
```

```python
#3. Replace the values which is more than 60 in the “age” column with the value of 50
for x in df.index:
    if df.loc[x,'age'] > 60:
        df.loc[x,'age'] = 50
print(df.to_string())
```

```python
#4. Find out the duplicates in the dataset and remove.
print(df.duplicated().to_string())
df.drop_duplicates(inplace = True)
print(df.to_string())
```

```python
#5. Find out the correlation between the data.
df.corr()
```

```python
#6. Draw a scatter plot between “age” and “bmi”
import matplotlib.pyplot as plt
df.plot(kind="scatter",x="age",y="bmi")
plt.show()
```
